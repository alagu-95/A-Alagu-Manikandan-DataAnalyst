#Pandas Series

```python
import numpy as np
import pandas as pd
```

1. Creating Pandas Series

```python
Marks = [95, 92, 89, 85, 80]
Custom_Indices = ['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']


marks = pd.Series(data = Marks, index = Custom_Indices)
marks
```

 2. Indexing and Slicing

```python
marks.iloc[0]
```

```python
marks.loc[:'Rank3']
```

```python
marks.iloc[2]
```

```python
marks[marks>90]
```

3. Manipulating Series

```python
marks['Rank1'] = 100
marks
```

```python
CGPA = marks/10
CGPA
```

```python
marks.drop('Rank5', inplace = True)
marks
```

#Pandas DataFrame

Creating Pandas DataFrame

```python
transactions = {
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
    'Region': ['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South', 'North'],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]

              }
```

3. Manipulating the DataFrame

```python
Transactions.iloc[1, -1] = 165
Transactions
```

```python
Transactions['Discount'] = (Transactions['Amount'] * 0.1)
Transactions
```

```python
Transactions.drop(8)
```

```python
Transactions = Transactions.drop('Discount', axis = 1)
```

4. Data Wrangling

```python
profit_data = {
    'TransactionID': [101, 102, 103, 105, 106, 107, 108, 109, 110],
    'Profit': [50, 20, 70, 40, 60, 80, 30, 90, 110]
}
profit_data = pd.DataFrame(profit_data)
profit_data
```

```python
Transactions.merge(profit_data, how = "left")
```

```python
Additional = {
              	'TransactionID': [111, 112],
                'ProductCategory': ['Toys', 'Books'],
                'Region': ['North', 'South'],
                'Amount': [220, 170]
}
```

```python
Additional = pd.DataFrame(Additional)
Additional
```

```python
Sales = pd.concat([Transactions, Additional], ignore_index= True)
Sales
```
