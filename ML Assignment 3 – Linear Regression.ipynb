```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Explore the Dataset

```python
plt.figure(figsize = (15,8))
sns.boxplot(mp)
plt.xticks(rotation = 45)
plt.show()
```

```python
min_value = mp.min()
Q1 = mp.quantile(0.25)
median = mp.median()
Q3 = mp.quantile(0.75)
max_value = mp.max()
print('min_value:\n', min_value)
print('Q1:\n', Q1)
print('median:\n', median)
print('Q3:\n', Q3)
print('max_value:\n', max_value)
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print('IQR:\n', IQR)
print('lower_bound:\n', lower_bound)
print('upper_bound:\n', upper_bound)
```

```python
mp = mp.copy()
```

```python
mp_outlier_removed = mp[(mp >= lower_bound) & (mp <= upper_bound)]
mp_outlier_removed
```

```python
mp_new = mp_outlier_removed.fillna(method= 'ffill')
```

```python
plt.figure(figsize = (15,8))
sns.boxplot(mp_new)
plt.xticks(rotation = 45)
plt.show()
```

```python
mp_new.describe().round( )
```

```python
correlation_matrix = mp_new.corr()
plt.figure(figsize= (10,6))
sns.heatmap(correlation_matrix, annot= True, linewidths= 0.4, fmt = '.1f')
plt.title('Correlation Matrix Heatmap')
plt.xticks(rotation = 90)
plt.show()
```

From the analysis

target variable have a positive strong correlation as well as negative correlation.

```python
top_features = correlation_matrix['Price'].abs().sort_values(ascending = False).round(1).head(4).index.tolist()
top_features
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 5))
for i, feature in enumerate(top_features):
    ax[i].scatter(mp_new['Price'], mp_new[feature], color='r')
    ax[i].set_title(f'{feature} vs Price')
    ax[i].set_ylabel(feature)
    ax[i].set_xlabel('Price')
plt.tight_layout()
plt.show()
```

#Prepare the Data

Feature Selection

```python
x = mp_new[['ram']]
y = mp_new['Price']

from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func= f_classif, k= 2)
x_new = selector.fit_transform(x, y)
feature_score = selector.scores_
print('feature score:', feature_score)
x.columns[selector.get_support(indices= True)]
print('selected Feature:', x.columns[selector.get_support(indices= True)] )
selected_mp = pd.DataFrame(x_new, columns = x.columns[selector.get_support(indices= True)])
print('Transformed dataset with selected mp:\n ', selected_mp)
```

Split the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.33, random_state= 30)
print(f'mp_new shape:, {mp_new.shape}' )
print(f'x_train shape: {x_train.shape}')
print(f'x_test shape: {x_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

#Build and Train the Model

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
m = model.coef_
c = model.intercept_
print(f'M: {m}')
print(f'C: {c}')
```

```python
y_pred = m*(x_test) + c
y_pred
```

```python
result = pd.concat([x_test, y_test, y_pred], axis = 1)
result.columns = ['ram', 'actual price', 'predicted price']
result.head()
```

```python
sns.lmplot(x= 'ram', y= 'actual price', data= result, line_kws={'color': 'red'})
plt.show()
```

```python
result['Residual'] = result['actual price'] - result['predicted price']
result.head()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print('r2_score=', r2_score(y_test, y_pred))
print('mean_absolute_error=', mean_absolute_error(y_test, y_pred))
print('mean_squared_error=', mean_squared_error(y_test, y_pred))
print('RSME=', mean_squared_error(y_test, y_pred)**0.5 )
```

#Conclude the Analysis

Based on the performance metrics


*  r2 score of the model is approx. 0.765, this indicates that 77% of the      variance in the target variable('Price').


*  The high r2 score suggests that the model explains a significant portion of the variance in Price, providing a strong prediction based on the selected feature.
*  But the prediction error range is high, indicating a improvement(outlier).
