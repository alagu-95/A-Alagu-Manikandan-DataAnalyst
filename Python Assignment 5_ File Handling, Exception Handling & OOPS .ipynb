#File Handling

Write

```python
with open("user_info.txt", "w") as file:
  file.write("Enter your Name:\n")
  file.write("Enter your Age:\n")
  file.write("Enter favorite color:\n")
```

Read

```python
with open("user_info.txt") as file:
  read = file.read()
  print(read)
```

Append

```python
with open("user_info.txt", "a") as file:
  file.write("Enter a favorite movie:")
```

Readlines

```python
with open("user_info.txt") as file:
  lines = file.readlines()
  for line in lines:
    print(line, end = "")

```

#Exception Handling

```python
def calc_temp():
  try:
    Celsius = float(input("Enter today temperature in celsius:"))
    Fahrenheit = (Celsius * 9/5) + 32
  except ValueError as e:
    print("the input should be a numeric value:",e)
  except Exception as e:
    print("an unexpected error occurred:", e)
  else:
    print(f"Calculated Fahrenheit:{Fahrenheit:.1f} degrees.")
  finally:
    print("The end of Calc_temp.")


```

#OOPS

```python
class Library:
  book_title = ""
  year_published = ""
  def __init__(self, book_title, year_published):
    self.book_title = book_title
    self.year_published = year_published
  def display_info(self):
    print(f"book_title:{self.book_title}\nyear_published:{self.year_published}")
```

```python
library = Library("Harry Potter And The Chamber Of Secrets", 1998 )
print(library.book_title)
```

```python
print(library.year_published)
```

```python
print(library.display_info())
```
