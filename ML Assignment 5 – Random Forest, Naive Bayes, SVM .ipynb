```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
df.head().round()
```

```python
le.fit(df['LaunchMonth'])
df['LaunchMonth'] = le.transform(df['LaunchMonth'])
df.head().round()
```

```python
le.fit(df['Country'])
df['Country'] = le.transform(df['Country'])
df.head().round()
```

```python
le.fit(df['Currency'])
df['Currency'] = le.transform(df['Currency'])
df.head().round()
```

```python
le.fit(df['VideoIncluded'])
df['VideoIncluded'] = le.transform(df['VideoIncluded'])
df.head().round()
```

```python
le.fit(df['CampaignID'])
df['CampaignID'] = le.transform(df['CampaignID'])
df.head().round()
```

#Data Splitting

```python
features = ['CampaignID','GoalAmount','RaisedAmount','DurationDays','NumBackers','Category','LaunchMonth','Country','Currency','OwnerExperience','VideoIncluded','SocialMediaPresence','NumUpdates']
target = 'IsSuccessful'

X = df[features]
Y = df[target]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size= 0.7,test_size = 0.2,random_state= 42)

print(f'Dataset Shape: {df.shape}')
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)
scaler.fit(X_test)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled)
print(X_test_scaled)
```

#Random Forest - Model Building and Evaluation

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=24)
model.fit(X_train, Y_train)
```

```python
n_estimators_values = [50, 100]
max_depth_values = [5, 10]

for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
        rf.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(n_estimators=100, random_state=24,max_depth=5)

print("Accuracy:")
print(accuracy_score(Y_test, Y_pred))
```

#Naive Bayes - Model Building and Evaluation

```python
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()
nb_model.fit(X_train_scaled,Y_train)
```

```python
Y_pred = nb_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
```

#Support Vector Machine (SVM) - Model Building and Evaluation

```python
from sklearn.svm import SVC
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train, Y_train)
```

```python
Y_pred = svm_classifier.predict(X_test_scaled)
```

  5.3. Reflect on the strengths and weaknesses of each model:

Random Forest:

Strengths:
Often achieves high accuracy.
Handles both categorical and numerical features well.
Can provide feature importance information.

Weaknesses:
Can be computationally expensive for large datasets.
Less interpretable compared to simpler models.

Naive Bayes:

Strengths:
Simple and easy to understand.
Works well with high-dimensional data.
Handles categorical features effectively.

Weaknesses:
Assumes independence between features, which may not always hold true.
Sensitive to irrelevant features.

SVM:

Strengths:
Effective in high-dimensional spaces.
Versatile due to different kernel options.
Can handle non-linearly separable data.

Weaknesses:
Sensitive to the choice of kernel and hyperparameters.
Can be computationally expensive for large datasets.
