#While loop & Control Statements :
(else, break, continue, pass)



Number Guessing Game

```python
import random
Num_Guessing = random.randint(1, 10)
Guess_Num = 0
Attempts = 3

while Attempts > 0:
    Guess_Num = int(input("Enter the your guess number:"))
    Attempts -= 1

    if Guess_Num < 1 or Guess_Num > 10 :
      print("Your guess number is out of range.")
      continue
    elif Guess_Num < Num_Guessing:
      print("Too High! Try again.")
    elif Guess_Num > Num_Guessing:
      print("Too low! Try again")

    elif Guess_Num == Num_Guessing:
      print("Congratulation! You guessed the correct number.")
      break

else:
  print("Better luck next time")



```
